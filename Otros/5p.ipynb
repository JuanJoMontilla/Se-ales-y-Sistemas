```python
%%writefile app.py

import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import streamlit as st

# Configuraci贸n inicial de la p谩gina
st.set_page_config(page_title="Parcial 2 - Fourier SyS", page_icon="", layout="wide")

# Barra lateral para seleccionar el punto del parcial y mostrar informaci贸n general
with st.sidebar:
    st.image("https://lh4.googleusercontent.com/proxy/WNtyuTbDjnnITJFxg1dlI63L0jfIMRf0CIKg75VavFd3ameUuokpEiXIZvafO0UbA3rGKkhjDZ2HFtRWcGiPIn7Syd37PqnCrQuXFNHguRRPYm__safRJi9Q", width=150)
    st.title("Parcial 2 - Fourier SyS")
    st.subheader("Universidad Nacional de Colombia - Sede Manizales")
    st.write("**Se帽ales y Sistemas - 2024-1**")
    st.write("**Integrantes:**")
    st.write("- Juan Felipe Posada Niasa")
    st.write("- Juan Jose Montilla Bastidas")

    # Men煤 desplegable para seleccionar el punto del parcial
    punto = st.selectbox("Selecciona el punto del parcial",
                         ("Punto 4: THD y Factor de Potencia",
                         "Punto 5: THD y Factor de Potencia"))
if punto == "Punto 4: THD y Factor de Potencia":
    st.write("### Punto 4: THD y Factor de Potencia")
    st.markdown("""
**1. Consulte en que consiste la distorcion total de armonicos(Total Harmonic Distortion-(THD)) y el factor de potencia en un circuito electrico.**

**THD:**
Consiste en que se parte de un sistema ideal en el que la funci贸n de transferencia es lineal e invariante en el tiempo.
Cuando una se帽al sinusoidal de frecuencia  pasa por un dispositivo no ideal y no lineal, se a帽ade contenido adicional en m煤ltiplos n (arm贸nicos) de la frecuencia original.
La THD es una medida de ese contenido de se帽al adicional no presente en la se帽al de entrada. Esta puede ser THD indic谩ndonos con respecto al voltaje o tambi茅n puede ser THDi, la cual es con respecto a la corriente.

**FP:**
Es la relaci贸n entre la potencia real y la potencia aparente en un circuito o sistema de distribuci贸n. La potencia real se mide en W o kW, y es la que utilizan los motores, las luces y otros dispositivos para producir trabajo 煤til. La potencia aparente se mide en VA o kVA, que es la suma vectorial de la potencia real (trabajo 煤til), la potencia reactiva (almacenada y liberada por inductores y condensadores y muestra el desplazamiento de fase entre las formas de onda de corriente y tensi贸n), y la potencia arm贸nica (la cual se pierde por distorsi贸n arm贸nica).
""")
    st.markdown("""
**2. 驴Como puede calcularse el THD desde la FFT?**

**Los pasos a seguir ser铆an:**

1) **Realizar la FFT:** Esto transforma la se帽al del dominio del tiempo al dominio de la frecuencia, permitiendo identificar las diferentes componentes de frecuencia, incluyendo la fundamental y los arm贸nicos.

2) **Identificar la frecuencia fundamental:** Una vez que se tiene el espectro de frecuencias, se debe identificar la frecuencia fundamental $V_1$, que es la frecuencia m谩s baja y representa la se帽al principal.

3) **Calcular las amplitudes de los arm贸nicos:** A partir del espectro de la FFT, se deben extraer las amplitudes de los arm贸nicos $V_n$, donde $n$ es el n煤mero del arm贸nico, comenzando desde el arm贸nico 2 y tomando los componentes de frecuencia que son m煤ltiplos enteros de la frecuencia fundamental.

4) **Aplicar la f贸rmula del THD:**

$$
THD = \\frac{\\sqrt{V_2^2 + V_3^2 + \\cdots + V_n^2}}{V_1}
$$

donde $V_n$ es el valor del $n$-茅simo arm贸nico de la se帽al de salida del circuito (en t茅rminos de tensi贸n), y $V_1$ es la componente fundamental.
""")
    st.markdown("""
**3. 驴Como puede calcularse la distorsi贸n del factor de potencia con base al THD?**

El factor de potencia de distorsi贸n es la relaci贸n entre la potencia real y la potencia aparente debido a la distorsi贸n arm贸nica total (THD). No se pueden a帽adir condensadores a un circuito para compensar el factor de potencia de distorsi贸n. La impedancia de los condensadores disminuye con la frecuencia. Por lo tanto, un condensador puede convertirse en un sumidero de arm贸nicos de alta frecuencia. Se utilizan tipos especiales de transformadores o filtros arm贸nicos sintonizados que constan de condensadores e inductores para corregir el factor de potencia de distorsi贸n.

El factor de potencia de distorsi贸n se calcula de la siguiente manera:

$$
PF_{THD} = \\sqrt{\\frac{1}{1 + (THD)^2}}
$$
""")
elif punto == "Punto 5: THD y Factor de Potencia":
    st.write("### Punto 5: THD y Factor de Potencia")
# Funci贸n para calcular el THD y la Distorsi贸n del Factor de Potencia
    def calculate_thd_and_pf(R_v):
        # Par谩metros
        Fo = 60  # Frecuencia de alimentaci贸n
        Fs = 30 * Fo  # Frecuencia de muestreo
        To = 1 / Fo  # Periodo fundamental
        num_samples = int(5 * To * Fs)  # N煤mero de muestras para 5 periodos de alimentaci贸n

        # Generar el vector de tiempo
        t = np.linspace(0, 5 * To, num_samples, endpoint=False)

        A = 120  # Amplitud de entrada

        # Onda senoidal
        in_o = A * np.sin(2 * np.pi * Fo * t)

        # Rectificador de onda completa
        rec_c = np.abs(np.sin(2 * np.pi * Fo * t))  # Simulaci贸n de rectificador de onda completa
        in_ = in_o * rec_c  # Se帽al de entrada rectificada

        # Definir la funci贸n de transferencia para R_v
        num = np.array([1])  # Numerador
        den = np.array([R_v, 1])  # Denominador
        G_n = sig.TransferFunction(num, den)

        # Evaluar salida ante entrada rectificada
        tout, out, _ = sig.lsim(G_n, in_, t)

        # Graficar se帽ales de entrada y salida
        fig, ax = plt.subplots()
        ax.plot(t, in_, label='Entrada In(t)', linestyle='--', color='blue')  # Entrada rectificada
        ax.plot(t, out, label=f'Salida Out(t) con R = {R_v} Ohmios', color='red')  # Salida del circuito
        ax.set_title('Se帽ales de Entrada y Salida del Circuito')
        ax.set_xlabel('Tiempo [s]')
        ax.set_ylabel('Amplitud')
        ax.legend()
        st.pyplot(fig)

        # Transformada de Fourier
        vfre = np.fft.rfftfreq(len(out), 1/Fs)  # Vector de frecuencias
        Xf = np.fft.rfft(out)  # FFT de la se帽al de salida

        # Graficar el espectro de frecuencias
        fig, ax = plt.subplots()
        ax.stem(vfre, np.abs(Xf)/len(out), use_line_collection=True)
        ax.set_title('Espectro de Frecuencias de la Se帽al de Salida')
        ax.set_xlabel('Frecuencia [Hz]')
        ax.set_ylabel('Magnitud |X(f)|')
        st.pyplot(fig)

        # C谩lculo de THD
        funda = np.abs(Xf[1])  # Amplitud de la frecuencia fundamental (primer arm贸nico)
        harmonics = np.abs(Xf[2:])  # Arm贸nicos (ignorando el componente DC)
        THD = np.sqrt(np.sum(harmonics**2)) / funda  # C谩lculo del THD

        # C谩lculo del Factor de Potencia (PF) basado en el THD
        PF_THD = np.sqrt(1 / (1 + THD**2))

        # Mostrar resultados
        st.write(f"**THD de la se帽al**: {THD:.4f}")
        st.write(f"**Factor de Potencia (PF) en funci贸n del THD**: {PF_THD:.4f}")

    # Interfaz en Streamlit
    st.title("C谩lculo de THD y Factor de Potencia para un Rectificador de Onda Completa con Carga Resistiva")

    # Slider para el valor de la resistencia R, ajustable entre 10 y 1000 Ohmios
    R_v = st.slider('Valor de la Resistencia (Ohmios)', min_value=10.0, max_value=1000.0, value=150.0, step=10.0)

    # Calcular y mostrar el resultado cuando el valor de R cambia
    calculate_thd_and_pf(R_v)
```
